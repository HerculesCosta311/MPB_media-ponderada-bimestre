```python
# Coleta da quantidade de tarefas e seus detalhes
quantidade_tarefas = int(input("Quantas tarefas foram aplicadas no bimestre? "))

tarefas = []
peso_total = 0

for i in range(1, quantidade_tarefas + 1):
    nome_tarefa = input(f"Nome da tarefa {i}: ")
    peso = float(input(f"Peso da tarefa '{nome_tarefa}': "))
    total_questoes = int(input(f"Quantas questões teve a tarefa '{nome_tarefa}'? "))
    peso_total += peso
    tarefas.append({
        "nome": nome_tarefa,
        "peso": peso,
        "questoes": total_questoes
    })
# Coleta da quantidade de alunos
quantidade_alunos = int(input("Quantos alunos deseja cadastrar? "))
alunos = []

# Cadastro de alunos
for aluno_index in range(quantidade_alunos):
    print(f"\n=== Cadastro do Aluno {aluno_index + 1} ===")
    nome = input("Nome do Aluno: ")
    sobrenome = input("Sobrenome do Aluno: ")
    nome_completo = f"{nome} {sobrenome}"

    total_acertos = 0
    total_questoes_respondidas = 0
    nota_total = 0

    tarefas_realizadas = []

    for tarefa in tarefas:
        while True:
            acertos = float(input(
                f"Acertos na tarefa '{tarefa['nome']}' (máximo {tarefa['questoes']}): "
            ))
            if 0 <= acertos <= tarefa["questoes"]:
                break
            print(f"Erro: acertos devem estar entre 0 e {tarefa['questoes']}.")

        nota = (acertos / tarefa["questoes"]) * tarefa["peso"]
        total_acertos += acertos
        total_questoes_respondidas += tarefa["questoes"]
        nota_total += nota

        tarefas_realizadas.append({
            "Tarefa": tarefa["nome"],
            "Acertos": acertos,
            "Questões": tarefa["questoes"],
            "Peso": tarefa["peso"],
            "Nota Proporcional": round(nota, 2)
        })

    media_final = nota_total
    aproveitamento = (total_acertos / total_questoes_respondidas) * 100
    situacao = "Aprovado" if media_final >= 6 else "Reprovado"

    alunos.append({
        "Nome Completo": nome_completo,
        "Tarefas": tarefas_realizadas,
        "Nota Final": round(media_final, 2),
        "Aproveitamento (%)": round(aproveitamento, 2),
        "Situação": situacao
    })

# Exibição dos resultados finais
print("\n=== RESULTADOS FINAIS ===")
for aluno in alunos:
    print(f"\nAluno: {aluno['Nome Completo']}")
    for tarefa in aluno["Tarefas"]:
        print(f"  - {tarefa['Tarefa']}: {tarefa['Acertos']}/{tarefa['Questões']} acertos | "
              f"Peso: {tarefa['Peso']} | Nota: {tarefa['Nota Proporcional']}")
    print(f"Nota Final: {aluno['Nota Final']} / {peso_total}")
    print(f"Aproveitamento: {aluno['Aproveitamento (%)']}%")
    print(f"Situação: {aluno['Situação']}")
```
